```python
import pandas as pd

df = pd.read_excel('/content/л.csv')

# Преобразование типов данных
df['Победы'] = pd.to_numeric(df['Победы'], errors='coerce').astype(float)  # Преобразуем 'победы' в float
df['Матчи'] = pd.to_numeric(df['Матчи'], errors='coerce')

# Обработка деления на 0
df['процент_побед'] = (df['Победы'] / df['Матчи']) * 100
df.loc[df['Матчи'] == 0, 'процент_побед'] = 0

# Замена NaN на 0 (для FutureWarning)
df['процент_побед'] = df['процент_побед'].fillna(0)

# Нахождение 10 команд с наибольшим процентом побед
top_10_teams = df.nlargest(10, 'процент_побед')

# Вывод результата
print("10 команд с наибольшим процентом побед:")
print(top_10_teams[['Команда', 'процент_побед']])

```

```python
import pandas as pd

df = pd.read_excel('/content/л.csv')

# Преобразование типов данных (если нужно)
df['Победы'] = pd.to_numeric(df['Победы'], errors='coerce').astype(float)
df['Матчи'] = pd.to_numeric(df['Матчи'], errors='coerce').astype(float)

# Создаем столбец "Эффективность"
df['Эффективность'] = df['Победы'] / df['Матчи']

# Обрабатываем деление на 0, если есть команды без игр
df.loc[df['Матчи'] == 0, 'Эффективность'] = 0

# Заменяем NaN на 0 в столбцах "Матчи" и "Победы"
df['Матчи'] = df['Матчи'].fillna(0)
df['Победы'] = df['Победы'].fillna(0)

# Сортируем по "Матчи" и "Эффективность"
df = df.sort_values(by=['Матчи', 'Эффективность'], ascending=[True, False])

# Выбираем топ-10 команд
top_10_teams = df.head(10)

# Выводим результат
print("Топ-10 команд, которые играли меньше всех, но выиграла больше всех (по эффективности):")
print(top_10_teams[['Команда', 'Матчи', 'Победы', 'Эффективность']])
```
